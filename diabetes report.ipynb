in this part we are importing pandas, NumPy and matplotlib libraries. 

```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

we import the csv file diabetes_dataset.csv

```
df=pd.read_csv('diabetes_dataset.csv')
df.head(100)
```

```
df.info()
df.shape
```

the report we can see that the data is sorted. hence no ne need to drop null values and convert datatypes

```
df['BMI'].describe()
```

```
plt.plot(df["BMI"], df["Age"], marker= '.', linestyle='',color='r')
plt.xlabel("BMI Level")
plt.ylabel("Age of participants")
plt.title("BMI vs Age")
plt.show()
```

we observe above the fact there is greater concentration in the middle of  X AXIS at  25 and 30 BMI level. for reference a BMI score of between 25 and 29.9 is classed as 'overweight'. do we have type 2 diabetes mainly in the dataset?

```
df["BMI"].mean()
```

and we are correct 

```
plt.plot(df["Age"], df["HbA1c"], marker= '.', linestyle='', color='r')
plt.xlabel("Age")
plt.ylabel("HbA1c")
plt.title("Age vs HbA1c")

plt.show()
```

our previous hypothesis may remain the same. we can observe that the younger peers have HBA1C  of 5 while the older peers have HBA1C greater or close to 6. 

```
df["HbA1c"].mean()
```

```
plt.bar(df['Age'], df['Glucose'])  
plt.xlabel("Age")
plt.ylabel("Glucose")
plt.title(" Glucose vs Age ")
plt.show()
```

in the graph above we can observe that the 20s age group have by far the most stable blood sugar at range of 120-130. we can also observe that the 60,70s and 80s also have increased blood sugar close to the range of the 200s. there could be several factors that contribute to this. such as metabolism activity(which declines with age).  participating in sporting activities(younger people tend to exercise more). which mostly explains why the younger group have and improved blood glucose compared to their older peers. 

lets not jump to conclusions and check the Triglycerides. maybe that could tell us more about the individuals and if they have type 2 diabetes or some other? 

```
df["Triglycerides"].mean()

plt.bar(df['Age'], df['Triglycerides'])  
plt.xlabel("Age")
plt.ylabel("Triglycerides")
plt.title(" Triglycerides vs Age ")
plt.show()
```

```
df["Triglycerides"].mean()
```

we are in the borderline high results range for Triglycerides. more Triglycerides levels suggests type 2 diabetes. could be also gestational diabetes? hmm?? lets first check the age? maybe we are dealing with seniors.
